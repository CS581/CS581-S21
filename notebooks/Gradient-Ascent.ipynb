# Gradient Ascent

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def f(x):
    return -2*(x**2) + 8*x + 10

def fp(x):
    return -4*x + 8

def fpp(x):
    return -4
```

```python
x_axis=np.linspace(-3, 7, 100)
fig, ax = plt.subplots()
ax.plot(x_axis, f(x_axis), linewidth=3, label='f', color='b')
ax.plot(x_axis, fp(x_axis), linewidth=3, label='fp', color='r')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.grid(True, which='both')

ax.legend()
```

```python
x=-2

lr1=0.2
lr2=0.3
lr3=0.4

lrp = 0.25

#print(x+lr1*fp(x))
#print(x+lr2*fp(x))
#print(x+lr3*fp(x))

print(x+lr1*fp(x))
print(f(x+lr1*fp(x)))
print()

print(x+lr2*fp(x))
print(f(x+lr2*fp(x)))
print()

print(x+lr3*fp(x))
print(f(x+lr3*fp(x)))
print()

print(x+lrp*fp(x))
print(f(x+lrp*fp(x)))
```

```python
lr=0.1
eps = 0.001
max_it = 100

x=8

all_xs = []
all_xs.append(x)

for i in range(max_it):    
    print("Iter %d:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(i, x, f(x), fp(x), lr*fp(x)))  
    x += lr*fp(x)
    all_xs.append(x)
    if abs(fp(x)) < eps:
        print("Final:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(x, f(x), fp(x), lr*fp(x)))   
        break
```

```python
x_axis=np.linspace(-3, 7, 100)
plt.plot(x_axis, f(x_axis))
plt.plot(all_xs, [f(x_) for x_ in all_xs], 'r+', markersize=20, markeredgewidth=2)
```

```python
x=-2
all_xs = []
all_xs.append(x)

for i in range(max_it):    
    print("Iter %d:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(i, x, f(x), fp(x), lr*fp(x)))  
    x += -fp(x)/fpp(x)
    all_xs.append(x)
    if abs(fp(x)) < eps:
        print("Final:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(x, f(x), fp(x), lr*fp(x)))   
        break
```

```python
def f(x):
    return -(x**5)-2*(x**4)+13*(x**3)+14*(x**2)-24*x

def fp(x):
    return -5*(x**4)-8*(x**3)+39*(x**2)+28*x-24

def fpp(x):
    return -20*(x**3)-24*(x**2)+78*x+28
```

```python
x_axis=np.linspace(-3.7, 2.7, 100)

fig, ax = plt.subplots()

ax.plot(x_axis, f(x_axis), linewidth=3, label='f', color='b')
ax.plot(x_axis, fp(x_axis), linewidth=3, label='fp', color='r')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.grid(True, which='both')

ax.legend()
```
