```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
from matplotlib import colors
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def perceptron(w, x):
    return np.sign(w[0]+ np.dot(w[1:], x))
```

```python
nx = ny = 200
x_min = y_min = -1.3
x_max = y_max = 1.3
xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
```

```python
D=np.c_[xx.ravel(), yy.ravel()]
```

```python
def and_perc(d):
    w = [-1, 1, 1]
    return perceptron(w, d)
```

```python
def or_perc(d):
    w = [1, 1, 1]
    return perceptron(w, d)
```

```python
#AND
plt.plot([-1], [-1], 'o', color='red', markersize=20)
plt.plot([-1], [1], 'o', color='red', markersize=20)
plt.plot([1], [-1], 'o', color='red', markersize=20)
plt.plot([1], [1], 'o', color='blue', markersize=20)

p=np.array([and_perc(d) for d in D])
p=p.reshape(xx.shape)


plt.contourf(xx, yy, p, cmap=plt.cm.RdBu, alpha=0.5)
plt.contour(xx, yy, p, [0], linewidths=2., colors='k')
```

```python
#OR
plt.plot([-1], [-1], 'o', color='red', markersize=20)
plt.plot([-1], [1], 'o', color='blue', markersize=20)
plt.plot([1], [-1], 'o', color='blue', markersize=20)
plt.plot([1], [1], 'o', color='blue', markersize=20)

p=np.array([or_perc(d) for d in D])
p=p.reshape(xx.shape)

plt.contourf(xx, yy, p, cmap=plt.cm.RdBu, alpha=0.5)
plt.contour(xx, yy, p, [0], linewidths=2., colors='k')
```

```python
def neg_perc(v):
    return -1*np.sign(v)
```

```python
def xor_nn(d):
    a = d[0]
    b = d[1]
    return or_perc([and_perc([a, neg_perc(b)]), and_perc([neg_perc(a), b])])
```

```python
#XOR
plt.plot([-1], [-1], 'o', color='red', markersize=20)
plt.plot([-1], [1], 'o', color='blue', markersize=20)
plt.plot([1], [-1], 'o', color='blue', markersize=20)
plt.plot([1], [1], 'o', color='red', markersize=20)

p=np.array([xor_nn(d) for d in D])
p=p.reshape(xx.shape)


plt.contourf(xx, yy, p, cmap=plt.cm.RdBu, alpha=0.5)
plt.contour(xx, yy, p, [0], linewidths=2., colors='k')
```
